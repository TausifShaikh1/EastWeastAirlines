```python
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler
scaler = MinMaxScaler()

# Find numerical columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Apply scaling to numeric columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df
```

```python
from sklearn.preprocessing import StandardScaler

# Select a subset of the data
df_subset = df[df.columns[1:6]]

# Apply Standard Scalar to the subset of data
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_subset), columns=df_subset.columns)
df_scaled
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Create linkage matrix
linked = linkage(df_scaled, 'ward')

# Create dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='level', p=5)
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns

# Create and fit an AgglomerativeClustering model
clustering = AgglomerativeClustering().fit(df_scaled)
df_scaled['cluster'] = clustering.labels_

# Plot the clusters
plt.figure(figsize=(10,15)) 
sns.scatterplot(data=df_scaled, x='Balance', y='Qual_miles', hue='cluster')

plt.title('Agglomerative Clustering')
plt.show()

# From the plot, we can observe that the agglomerative clustering algorithm separates the dataset into reasonable clusters.
# Another insight would be gaining an understanding of the patterns in the dataset, numberOf clusters, decision boundaries, etc.
# Here we can see that most of the balance that qual_miles have is concetrated from 0 to 4, and stretches from 0 till 10.0.
# Later the data is way more distributed in qual_miles and balance
```
